# list_plot

The `list_plot` function simplifies plotting of one-dimensional data.

    list_plot([y_1, y_2, ...]) plots the points (1, y_1), (2, y_2), ...
    list_plot([[x_1, y_1], [x_2, y_2], ...]) plots the points (x_1, y_1), (x_2, y_2), ...
    list_plot([data_1, data_2, ...]) combines several plots.

```python
from neuralprocesses.utils.plotting import list_plot
from neuralprocesses.utils.plotting import Color
```

## Basic examples

Plot a couple of y-values against their `index + 1`.

```python
list_plot(range(10))
```

Join the points with a line and change the color.

```python
list_plot(range(10), joined=True, plot_style=Color((0.25, 0.25, 0.9)))
```

Specify the steping if x-values are not given.

```python
list_plot([1, 6, 5], step=(2, 8, 3))
```

Provide x-values explicitly.

```python
list_plot([[0, 2], [10, 4], [11, 3], [12, 5], [30, 3]], axes_label=("x in seconds", "y in kilograms"))
```

Combine multiple datasets and style them individually.

```python
list_plot([
        [[0, 2], [10, 4], [11, 3], [12, 5], [30, 3]],
        [[1, 1.3], [2, 1], [17, 2]],
        [[0, 3], [10, 3.3], [11, 3.6], [12, 4], [30, 2]]
    ], axes_label=("x", "y"), joined=[False, True, True], mesh=[True, True, False])
```
