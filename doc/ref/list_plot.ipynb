# list_plot

The `list_plot` function simplifies plotting of one-dimensional data.

    list_plot([y_1, y_2, ...]) plots the points (1, y_1), (2, y_2), ...
    list_plot([[x_1, y_1], [x_2, y_2], ...]) plots the points (x_1, y_1), (x_2, y_2), ...
    list_plot([data_1, data_2, ...]) combines several plots.

```python
from neuralprocesses.utils.plotting import list_plot
from neuralprocesses.utils.plotting import Color
```

## Basic examples

Plot a couple of y-values against their `index + 1`.

```python
list_plot(range(10))
```

Join the points with a line and change the color.

```python
list_plot(range(10), joined=True, plot_style=Color((0.25, 0.25, 0.9)))
```

Specify the steping if x-values are not given.

```python
list_plot([1, 6, 5, 3, 4, 7, 8], step=(2, 2+6+6, 2), plot_style="x")
```

Provide x-values explicitly.

```python
list_plot([[0, 2], [10, 2.1], [11, 2.5], [12, 2.7], [30, 3]], axes_label=("x in seconds", "y in kilograms"))
```

Combine multiple datasets and style them individually.

```python
list_plot([
        [[0, 2], [10, 4], [11, 3], [12, 5], [30, 3]],
        [[1, 1.3], [2, 1], [17, 2]],
        [[0, 3], [10, 3.3], [11, 3.6], [12, 4], [30, 2]]
    ], axes_label=("x", "y"), joined=[False, True, True], mesh=[True, True, False])
```

Fill the area between curves 1 (blue) and 0 (red).

```python
list_plot([
        [[0.0, 1.0], [0.1, 1.01], [0.2, 1.04], [0.3, 1.09], [0.4, 1.16], [0.5, 1.25], [0.6, 1.36],
         [0.7, 1.49], [0.8, 1.64], [0.9, 1.81]],
        [[0.0, 0.0], [0.1, 0.1], [0.2, 0.2], [0.3, 0.3], [0.4, 0.4], [0.5, 0.5], [0.6, 0.6],
         [0.7, 0.7], [0.8, 0.8], [0.9, 0.9]]
    ], axes_label=("x", "y"), joined=True, filling=[[1,0]])
```

Adjust colors and other settings for each graph with `plot_style`.

```python
f1 = [[4 * x, 1 + 0.2 * x**2] for x in range(20)]
f2 = [[x, 2 * x - 0.3 * (x*0.1)**2] for x in range(80)]

list_plot(
    [f1,f2],
    plot_style=[
        {"Color": Color((0.6,0.2,0.2)), "Marker": "x", "PointSize": 100},
        Color((0.1,0.2,0.6))              
    ]
)
```
